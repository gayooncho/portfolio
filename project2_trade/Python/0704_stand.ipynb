```python
import pandas as pd
import os

files = os.listdir('./map/')

file_name = []
# files 리스트 요소를 ,로 스플릿한 후 앞 요소만 가져오기
for i in range(len(files)):
    file_name.append(files[i].split('.')[0])
    globals()[files[i].split('.')[0]] = pd.read_csv(f'./map/{files[i]}')
    
```

```python
df_list = []

for df in file_name:      
    df = globals()[df]
    df_list.append(df)
```

```python
# 불필요한 열들제거
# unnamed 0.1, HS, unnamed:12, unnamed: 0, HS4, INDEX
drop_list = ['Unnamed: 0', 'no', 'no.1']

for df in df_list:
    df.drop(drop_list, axis = 'columns', inplace = True)
    # print(type(df))
# usa_imported.drop(drop_list, axis = 'columns', inplace = True)
```

```python
# 1. 열 추가 main country (concat 사용)
# 2. 열에 각 나라 이름 추가
```

```python
# 새로운 열 추가할 df 만들기
col_country = pd.DataFrame({'Main Country' : []})

# main country 에 넣을 나라이름
country_name = ['Australia', 'China', 'Germany', 'India', 'Japan', 'Korea Republic of', 'Russian Faderation',
                'Saudi Arabia', 'Singapore', 'Taipei Chinese', 'United States of America', 'Viet Nam']

df_main_country = []
for i in range(len(df_list)):
    df = df_list[i]
    df_new = pd.concat([df, col_country], axis = 1)
    df_new.loc[df_new['Main Country'] != df_new['Main Country'], 'Main Country'] = country_name[i]
    df_main_country.append(df_new)
```

```python
df_main_country[1]
```

```python
df_all = pd.DataFrame({'BEC' : [''], 'USD' : [''],
                       'Contry' : [''],
                       'Year' : [''],
                       'Main Country' : ['']})

for i in range(len(df_main_country)):
    df = df_main_country[i]
    df_all = pd.concat([df_all, df], axis=0)
```

```python
df_all.to_csv('./df_between_map_all.csv')
```
