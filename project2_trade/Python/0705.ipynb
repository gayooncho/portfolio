```python
import pandas as pd
import os

files = os.listdir('./import_export_year/import_year/')

file_name = []
# files 리스트 요소를 ,로 스플릿한 후 앞 요소만 가져오기
for i in range(len(files)):
    file_name.append(files[i].split('.')[0])
    globals()[files[i].split('.')[0]] = pd.read_csv(f'./import_export_year/import_year/{files[i]}')
```

```python
### export 나라들
df_list = []

for df in file_name:      
    df = globals()[df]
    df_list.append(df)
```

```python
drop_list = ['Unnamed: 0']

for df in df_list:
    df.drop(drop_list, axis = 'columns', inplace = True)
```

```python
df_list[0]
```

```python
# 새로운 열 추가할 df 만들기
col_imex = pd.DataFrame({'import, export' : []})

# main country 에 넣을 나라이름
imex = ['import']

df_main_imex = []
for i in range(len(df_list)):
    df = df_list[i]
    df_new = pd.concat([df, col_imex], axis = 1)
    df_new.loc[df_new['import, export'] != df_new['import, export'], 'import, export'] = imex[0]
    df_main_imex.append(df_new)
```

```python
df_import = pd.DataFrame({'BEC' : [''], 'USD' : [''],
                       'Country' : [''],
                       'Year' : [''],
                       'import, export' : ['']})

for i in range(len(df_main_imex)):
    df = df_main_imex[i]
    df_import = pd.concat([df_import, df], axis=0)
```
