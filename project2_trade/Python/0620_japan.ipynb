```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 일본 수출 물품 리스트 불러오기
japan_exported_list = pd.read_table('C:\project2\Trade_map_List_of_products_exported_and_imported\Trade_Map_-_List_of_products_exported_by_Japan.txt', sep = '\t')
japan_exported_list.rename(columns={'Code':'HS'}, inplace=True)
japan_exported_list.head()
```

```python
# hs, bec 파일 불러오기
hs17_bec5 = pd.read_csv('C:\project2\BEC5Correlations\HS17BEC5.csv')

# hs, bec 코드 불러오기
# hs code, 이름, bec 코드 추출
hs_bec = hs17_bec5[['HS6', 'BEC5Code1']]

# hs code -> code 이름변경
hs_bec.rename(columns={'HS6':'HS', 'BEC5Code1':'BEC'}, inplace=True)
hs_bec.head()
```

```python
# 데이터 합치기
# code에 total 000000 으로 바꿔주기
japan_exported_list.loc[0, 'HS'] = 000000

japan_exported_list['HS'] = pd.to_numeric(japan_exported_list['HS'])

# korea_list.info()

# hs_bec.dtypes

japan_exported_list = pd.merge(japan_exported_list, hs_bec, how='left', on='HS')
japan_exported_list.head()
```
