```python
# 폴더 내에 파일 리스트를 files에 할당
import os
import pandas as pd
files = os.listdir('C:/project2/Python/each_country_export_import/export/')

file_name = []
# files 리스트 요소를 ,로 스플릿한 후 앞 요소만 가져오기
for i in range(len(files)):
    file_name.append(files[i].split('.')[0])
    globals()[files[i].split('.')[0]] = pd.read_csv(f'C:/project2/Python/each_country_export_import/export/{files[i]}')
    
```

```python
df_list = []

for df in file_name:      
    df = globals()[df]
    df_list.append(df)
```

```python
# 불필요한 열들제거
# unnamed 0.1, HS, unnamed:12, unnamed: 0, HS4, INDEX
drop_list = ['Product label', 'HS', 'Unnamed: 12','Unnamed: 0', 'Unnamed: 0', 'HS4', 'index']

for df in df_list:
    df.drop(drop_list, axis = 'columns', inplace = True)
    # print(type(df))
# vietnam_imported.drop(drop_list, axis = 'columns', inplace = True)
```

```python
# BEC 별 합산
df_group = []

for df in df_list:
    data = pd.DataFrame(df.groupby('BEC').sum())
    df_group.append(data)
    # print(type(df))
```
