```python
# 폴더 내에 파일 리스트를 files에 할당
import os
import pandas as pd
files = os.listdir('C:/project2/Python/map_year/')

file_name = []
# files 리스트 요소를 ,로 스플릿한 후 앞 요소만 가져오기
for i in range(len(files)):
    file_name.append(files[i].split('.')[0])
    globals()[files[i].split('.')[0]] = pd.read_csv(f'C:/project2/Python/map_year/{files[i]}')
    
    
```

```python
df_list = []

for df in file_name:      
    df = globals()[df]
    df_list.append(df)
```

```python
australia = pd.DataFrame(df_list[0])    #0 ~ 9
china = pd.DataFrame(df_list[10])       #10 ~
germany = pd.DataFrame(df_list[20])     
india = pd.DataFrame(df_list[30])
japan = pd.DataFrame(df_list[40])
korea = pd.DataFrame(df_list[50])
russian = pd.DataFrame(df_list[60])
saudi = pd.DataFrame(df_list[70])
singapore = pd.DataFrame(df_list[80])
taipei = pd.DataFrame(df_list[90])
usa = pd.DataFrame(df_list[100])
vietnam = pd.DataFrame(df_list[110])
```
