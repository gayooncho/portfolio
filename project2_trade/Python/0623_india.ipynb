```python
import numpy as np
import pandas as pd
```

```python
# hs / bec 파일 불러오기, excel 파일의 두번째 sheet
hs_bec = pd.read_excel('C:\project2\BEC5Correlations\HS2012-17-BEC5 -- 08 Nov 2018.xlsx'
                       , sheet_name = "HS17BEC5")

# hs4, hs6, bec, bec processing, bec specitication 목록만 별도로 추출
hs_bec = hs_bec[['HS4','HS6','BEC5Code1', 'BEC5Processing', 'BEC5Specification']]

# hs code -> code 이름변경
hs_bec.rename(columns={'HS4':'HS4', 'HS6':'HS', 'BEC5Code1':'BEC', }, inplace=True)
```

```python
hs4_bec = hs_bec[['HS4','BEC']]

# hs4, bec코드 정리
hs4_bec['HS4'] = hs4_bec['HS4'].astype('int')


hs4_bec.dropna(axis=0, inplace= True)
# hs4_count = hs4_bec['HS4'].value_counts()
# hs4_count = pd.DataFrame(hs4_count)

# hs4_count.reset_index(inplace=True)
# hs4_count.columns = ['HS4', 'HS4_COUNT']
# hs4_bec.groupby('HS4').count()

hs4_bec['HS4'] = hs4_bec['HS4'].astype('int')
hs4_bec['BEC'] = hs4_bec['BEC'].astype('float')
```

```python
hs4 = hs4_bec['HS4']

hs4.drop_duplicates(inplace=True)

hs4 = hs4.to_frame()

hs4.reset_index(inplace=True)
hs4.count()
```

```python
# hs4에 대하여 bec 코드가 없는 행
# bec 값에 0

hs4_bec.loc[3513,'HS4'] = 0
hs4_bec.loc[3542,'HS4'] = 0
```

```python
# hs4와 bec 사전 만들기

from collections import defaultdict
hs4_dict = defaultdict(list)
    
hs4_bec.loc[0, 'HS4']
for i in range(len(hs4)) :
    hs4Temp = hs4.loc[i, 'HS4']
    for j in range(len(hs4_bec)):
        hsbecTemp = hs4_bec.loc[j, 'HS4']
        if hs4Temp == hsbecTemp and hs4_bec.loc[j, 'BEC'] not in hs4_dict[hs4Temp]:
            hs4_dict[hs4Temp].append(hs4_bec.loc[j, 'BEC'])
        pass
```

```python
# korea exported 파일 불러오기
japan_exported_list = pd.read_table('C:\project2\Trade_map_List_of_products_exported_and_imported\Trade_Map_-_List_of_products_exported_by_India.txt', sep = '\t')

# code -> hs로 컬럼 이름 변경
japan_exported_list.rename(columns={'Code':'HS'}, inplace=True)
japan_exported_list.head()
```

```python
#   1) code 열에 total -> 0으로 변경
# korea_exported_list.loc[0, 'HS'] = 000000
# 제거
japan_exported_list = japan_exported_list.drop(index = 0, axis = 0)

#   2) japan 데이터 프레임에 hs 코드 데이터 타입 형변환
japan_exported_list['HS'] = pd.to_numeric(japan_exported_list['HS'])

#   2) merge를 사용해 hs코드 기준으로 데이터 프레임 병합
japan_exported_list = pd.merge(japan_exported_list, hs_bec, how='left', on='HS')

#   2-2 hs코드 잘라서 hs4에 넣어주기
japan_exported_list['HS'] = japan_exported_list['HS'].astype('str')
japan_exported_list['HS4'] = japan_exported_list['HS'].str.slice(0, 4)

#   3) 병합된 데이터 프레임에 결측값 확인
japan_exported_list.isnull().sum()

#

no_exported_id = japan_exported_list[(japan_exported_list['Exported value in 2017'] == 0)
                    & (japan_exported_list['Exported value in 2018'] == 0)
                    & (japan_exported_list['Exported value in 2019'] == 0)
                    & (japan_exported_list['Exported value in 2020'] == 0)
                    & (japan_exported_list['Exported value in 2021'] == 0)].index

# 거래품목 없는 인덱스 제거
japan_exported_list = japan_exported_list.drop(no_exported_id)
japan_exported_list.isnull().sum()
```

```python
add_BEC = japan_exported_list[japan_exported_list['BEC'].isnull()]
add_BEC['HS'] = add_BEC['HS'].astype('float')
add_BEC

# add_index = add_BEC.index
# add_index= list(add_index)

hs4_count = hs4_bec.groupby('HS4').count()
add_BEC.head()

hs4_hs_bec = hs_bec[['HS','BEC']]
add_BEC.drop(['BEC'], axis=1, inplace=True)


add_BEC = pd.merge(add_BEC, hs4_hs_bec, how='left', on='HS')
add_BEC.head()
# add_BEC.isna().count()
```

```python
### bec 코드 부여
# dic 에서 하나씩 넣어주기?
for i in range(len(add_BEC)):
    com_key = int(add_BEC.loc[i, 'HS4'])
    for key, value in hs4_dict.items():
        if key == com_key:
            add_BEC.loc[i, 'BEC'] = hs4_dict[key][0]

add_BEC
```
