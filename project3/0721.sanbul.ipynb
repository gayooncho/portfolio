```python
sanbul_24 = pd.read_excel('C:\project22\sanbul.xls'
                       , sheet_name = "Sheet1")

sanbul_19 = pd.read_csv('C:\project22\python\sanbul_90_14.csv')
```

```python
# 열 이름이 되야할 데이터 추출
index_fix = sanbul_24.loc[1]

# 열이름 변경
sanbul_24.columns = index_fix

# 필요없는 행 제거
sanbul_24 = sanbul_24.drop([0, 1])

sanbul_24['tm'] = sanbul_24['발생일시_년'] + '-' + sanbul_24['발생일시_월'] + '-' + sanbul_24['발생일시_일']
```

```python
sanbul = pd.concat([sanbul_19, sanbul_24],  axis = 0)
# sanbul = pd.merge(sanbul_19, sanbul_24, )
```
