```python
import pandas as pd
```

```python
sanbul = pd.read_excel('C:\project22\sanbul.xls'
                       , sheet_name = "Sheet1")
```

```python
sanbul.head()
```

```python
# 열 이름이 되야할 데이터 추출
index_fix = sanbul.loc[1]

# 열이름 변경
sanbul.columns = index_fix

# 필요없는 행 제거
sanbul = sanbul.drop([0, 1])
```

```python
# 년 월 일 합치기?
# 날짜 데이터와 형태 같게 만들기 yyyy-mm-dd
sanbul['tm'] = sanbul['발생일시_년'] + '-' + sanbul['발생일시_월'] + '-' + sanbul['발생일시_일']
```

```python
# 필요 한 열 추출
sanbul_data = sanbul[['tm', '발생장소_시도', '발생장소_시군구', '발생원인_구분', '피해면적_합계']]
sanbul_data.tail()
```

```python
# 결츨치 확인 = 없음
sanbul_data.isnull().sum()
```

```python
sanbul_data.to_csv('./sanbul_data.csv', encoding= 'euc-kr')
```
