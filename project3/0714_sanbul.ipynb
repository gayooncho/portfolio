```python
import pandas as pd
```

```python
sanbul = pd.read_excel('C:\project22\sanbul.xls'
                       , sheet_name = "Sheet1")
```

```python
# 열 이름이 되야할 데이터 추출
index_fix = sanbul.loc[1]

# 열이름 변경
sanbul.columns = index_fix

# 필요없는 행 제거
sanbul = sanbul.drop([0, 1])

# 년 월 일 합치기?
# 날짜 데이터와 형태 같게 만들기 yyyy-mm-dd
sanbul['tm'] = sanbul['발생일시_년'] + '-' + sanbul['발생일시_월'] + '-' + sanbul['발생일시_일']


# 발생장소_시도 = 강원, 경북, 울진만 추출

condition = (sanbul.발생장소_시도 == '강원') | (sanbul.발생장소_시도 == '경북') | (sanbul.발생장소_시도 == '울산')
sanbul_region = sanbul[condition]
```

```python
# 필요 한 열 추출
sanbul_data = sanbul_region[['tm', '발생일시_월','발생일시_시간','발생장소_시도', '발생장소_시군구', '발생원인_구분', '피해면적_합계']]
# 결츨치 확인 = 없음
sanbul_data.isnull().sum()
```

```python
sanbul.groupby('발생장소_시도').count()
```

```python
# 울산 시군구 통일
sanbul_data.loc[sanbul_data['발생장소_시도'] == '울산', '발생장소_시군구'] = '울산'
```

```python
# 발생일시_시간 시, 분 나누기
time_split = sanbul_data["발생일시_시간"].str.split(":")

sanbul_data["hour"] = time_split.str.get(0)
sanbul_data["min"] = time_split.str.get(1)

# 비교를 위해 형변환
sanbul_data["hour"] = sanbul_data["hour"].astype("int")
```

```python
# 산불 시간대 계산 (산불 발생 시를 조사한 후 시간대별로 나눔)
def func(x) :
    if x >= 0 and x < 3 :
        return "0 ~ 2"
    elif x >= 3 and x < 6:
        return "3 ~ 5"
    elif x >= 6 and x < 9:
        return "6 ~ 8"
    elif x >= 9 and x < 12:
        return "9 ~ 11"
    elif x >= 12 and x < 15:
        return "12 ~ 14"
    elif x >= 15 and x < 18:
        return "15 ~ 17"
    elif x >= 18 and x < 21:
        return "18 ~ 20"
    else :
        return "21 ~ 23"
sanbul_data["time"] = sanbul_data["hour"].apply(lambda x : func(x))
```

```python
time_groups = sanbul_data.groupby('time')

# 시간대별 산불 발생 건수 확인
# 12 ~ 14 380
# 15 ~ 17 251
# 9 ~ 11 138
# 18 ~ 20 90
time_groups.count()
```

```python
condition = (sanbul.발생장소_시도 == '강원') | (sanbul.발생장소_시도 == '경북') | (sanbul.발생장소_시도 == '울산')
sanbul_region = sanbul[condition]
```

```python
rank_df = sanbul_data

rank_df['피해면적_합계'] = rank_df['피해면적_합계'].astype(float)

rank_df['rank'] = rank_df['피해면적_합계'].rank(ascending = False)

# rank_df[(rank_df['rank'] < 20)]
rank_df = rank_df.sort_values(by = 'rank', ignore_index = True)

rank_df
```

```python
# 피해면적 비율 계산하기

# 전체 피해면적 합산
total_area = rank_df['피해면적_합계'].sum()

# 피해면적/total_area 비율 계산 100%로 계산
rank_df['radio'] = rank_df['피해면적_합계']/total_area * 100

# 피해면적 누계 계산 # 누계합 함수 cumsum
rank_df['cumulative'] = rank_df['피해면적_합계'].cumsum()

rank_df['cumulative'] = rank_df['radio'].cumsum()
```
