```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sanbul = pd.read_excel('C:\project22\sanbul.xls'
                       , sheet_name = "Sheet1")

# 열 이름이 되야할 데이터 추출
index_fix = sanbul.loc[1]

# 열이름 변경
sanbul.columns = index_fix

# 필요없는 행 제거
sanbul = sanbul.drop([0, 1])

# 년 월 일 합치기?
# 날짜 데이터와 형태 같게 만들기 yyyy-mm-dd
sanbul['tm'] = sanbul['발생일시_년'] + '-' + sanbul['발생일시_월'] + '-' + sanbul['발생일시_일']

# 발생장소_시도 = 강원, 경북, 울진만 추출

# condition = (sanbul.발생장소_시도 == '경북') |(sanbul.발생장소_시도 == '강원')
# sanbul_region = sanbul[condition]
sanbul.reset_index(drop = True, inplace = True)
```

```python
# sanbul['발생장소_시도'] = sanbul['발생장소_시도'] + '도'

for i in range(len(sanbul)):
    sanbul_do = sanbul.loc[i, '발생장소_시도']
    if sanbul_do == '강원':
        sanbul_si_do = '강원도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '경북':
        sanbul_si_do = '경상북도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '경남':
        sanbul_si_do = '경상남도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '전북':
        sanbul_si_do = '전라북도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '전남':
        sanbul_si_do = '전라남도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '충북':
        sanbul_si_do = '충청북도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '충남':
        sanbul_si_do = '충청남도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '경기':
        sanbul_si_do = '경기도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '제주':
        sanbul_si_do = '제주도'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '부산':
        sanbul_si_do = '부산광역시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '서울':
        sanbul_si_do = '서울광역시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '광주':
        sanbul_si_do = '광주광역시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '세종':
        sanbul_si_do = '세종특별시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '울산':
        sanbul_si_do = '울산광역시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
    elif  sanbul_do == '인천':
        sanbul_si_do = '인천광역시'
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
        
    else:
        sanbul_si_do = sanbul_do
        
        sanbul.loc[i, '발생장소_시도'] = sanbul_si_do
        
```
