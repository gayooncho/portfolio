```python
import pandas as pd
```

```python
sanbul = pd.read_excel('C:\project22\sanbul.xls'
                       , sheet_name = "Sheet1")
```

```python
# 열 이름이 되야할 데이터 추출
index_fix = sanbul.loc[1]

# 열이름 변경
sanbul.columns = index_fix

# 필요없는 행 제거
sanbul = sanbul.drop([0, 1])
```

```python
# 년 월 일 합치기?
# 날짜 데이터와 형태 같게 만들기 yyyy-mm-dd
sanbul['tm'] = sanbul['발생일시_년'] + '-' + sanbul['발생일시_월'] + '-' + sanbul['발생일시_일']
```

```python
# 발생장소_시도 = 강원, 경북, 울진만 추출

condition = (sanbul.발생장소_시도 == '강원') | (sanbul.발생장소_시도 == '경북') | (sanbul.발생장소_시도 == '울산')

# 발생장소_시도 = 이름변경
sanbul.loc[sanbul['발생장소_시도'] == '울산', '발생장소_시도'] = '울산광역시'
sanbul.loc[sanbul['발생장소_시도'] == '강원', '발생장소_시도'] = '강원도'
sanbul.loc[sanbul['발생장소_시도'] == '경북', '발생장소_시도'] = '경상북도'

sanbul_region = sanbul[condition]
```

```python
# 필요 한 열 추출
sanbul_data = sanbul_region[['tm', '발생일시_시간','발생장소_시도', '발생장소_시군구', '발생장소_읍면', '발생장소_동리','발생원인_구분', '피해면적_합계']]

sanbul_data.tail()
```

```python
# 지역코드 넣기 위한 열
sanbul_data['region'] = sanbul_data['발생장소_시도'] + sanbul_data['발생장소_시군구'] + sanbul_data['발생장소_읍면']
sanbul_data
```

```python
code = pd.read_excel('C:/project22/adm_code.xls'
                       , sheet_name = "2021년6월")
```

```python
# 열이름 변경
code.columns = code.loc[0]

code.drop([0], inplace=True)
code.reset_index(inplace= True, drop=True)
```

```python
code
code['code'] = code['시도코드'] + code['시군구코드'] + code['읍면동코드']

# 명칭에 마지막 제거 / 문자열 슬라이싱

for i in range(len(code)):
    si = code.loc[i, '시도명칭']
    gu = code.loc[i, '시군구명칭'][:-1]
    dong = code.loc[i, '읍면동명칭'][:-1]
    code.loc[i, 'region'] = si + gu + dong
```
